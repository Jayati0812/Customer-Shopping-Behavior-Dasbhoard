```python
data.columns=data.columns.str.lower()
data.columns=data.columns.str.replace(' ',"_")
```

```python
data=data.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
(data['discount_applied']==data['promo_code_used']).all()
```

```python
data.drop('promo_code_used',axis=1)
```

```python
data.isnull().sum()
```

```python
data['review_rating']=data.groupby('category')['review_rating'].transform(lambda x: x.fillna(x.median()))
```

```python
labels=['Young Adult','Adult','Middle-aged','Senior']
data['age_group']=pd.qcut(data['age'],q=4,labels=labels)
```

```python
data[['age','age_group']]
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
data['purchase_frequency_days']=data['frequency_of_purchases'].map(frequency_mapping)
```

```python
data[['frequency_of_purchases','purchase_frequency_days']]
```
